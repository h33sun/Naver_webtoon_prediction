```python
from bs4 import BeautifulSoup
import requests
import re
from datetime import datetime
import pandas as pd
from selenium import webdriver
```

```python
# webtoon_episode 크롤링
# webtoon_list 가 불러와져 있어야 사용가능
data = []

selector_wrt_nm = '#content > div.section_cont.wide > div.comicinfo > div.detail > h2 > .wrt_nm'
selector_title = '#comicRemocon > div.remote_cont > a'
selector_episode = '#content > div.section_cont.wide > div.tit_area > div.view > h3'
selector_title_id = '#content > div.section_cont.wide > div.comicinfo > div.detail > ul > li:nth-child(1) > a'
selector_points = '#topPointTotalNumber'
selector_date = '#content > div.section_cont.wide > div.tit_area > div.vote_lst > dl.rt > dd:nth-child(2)'
selector_view = '#content > div.section_cont.wide > div.tit_area > div.vote_lst > dl.rt > dd:nth-child(4)'
selector_person = '#topTotalStarPoint > span.pointTotalPerson > em'

for i in ['title_id']:
    url = f'https://comic.naver.com/bestChallenge/list.nhn?titleId={i}'
    response1 = requests.get(url)
    dom1 = BeautifulSoup(response1.content, 'html.parser')

    for i in range(len(dom1.select('td.title > a'))):
        urls = 'https://comic.naver.com' + dom1.select('td.title > a')[i].get('href')
        print('https://comic.naver.com' + dom1.select('td.title > a')[i].get('href'))
        response2 = requests.get(urls)
        print(response2)
        dom = BeautifulSoup(response2.content, 'html.parser')
        result = dom.select_one(selector_title_id).get('href')
        print(result)
        results = re.findall('[0-9]', result)
        data.append({
            'title': dom.select_one(selector_title).text,
            'wrt_nm': dom.select_one(selector_wrt_nm).text.strip(),
            'episode': dom.select_one(selector_episode).text,
            'title_id': ''.join(results),
            'points': float(dom.select_one(selector_points).text),
            'date': dom.select_one(selector_date).text.replace('.', '-'),
            'view': int(dom.select_one(selector_view).text),
            'person': int(dom.select_one(selector_person).text)
        })
df = pd.DataFrame(data)
df
```

```python
# csv파일로 저장
df.to_csv("webtoon_episode.csv", encoding='utf-8-sig',index=False)
```
